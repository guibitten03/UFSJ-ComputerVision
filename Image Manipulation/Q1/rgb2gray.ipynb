```python
import matplotlib.pyplot as plt
from skimage import io

from skimage.color import rgb2gray
```

```python
lenna_img = io.imread('../Pictures/Lenna.png')

lenna_gray = rgb2gray(lenna_img)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(lenna_img)
ax[0].set_title("Original")
ax[1].imshow(lenna_gray, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
peppers_img = io.imread('../Pictures/peppers.jpg')

peppers_gray = rgb2gray(peppers_img)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(peppers_img)
ax[0].set_title("Original")
ax[1].imshow(peppers_gray, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```
